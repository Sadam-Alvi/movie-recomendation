```python
import pandas as pd
import numpy as np
```

```python
movies = pd.read_csv("tmdb_5000_movies.csv")
credit = pd.read_csv("tmdb_movie_credits.csv") 
movies
```

```python
credit
```

```python
df = pd.merge(movies, credit, on='title', how='inner')
df
```

```python
df.columns
```

```python
df = df[['genres','id','keywords','original_title','overview', 'release_date','cast', 'crew']]

```

```python
df.isna().sum()
df.dropna(inplace=True)
```

```python
df.info()
```

```python
df.drop_duplicates(inplace=True)
df.duplicated().sum()
```

```python
df["genres"].head(1)
```

```python
import ast
def convert(obj):
    list1 = []
    for i in ast.literal_eval(obj):
         list1.append(i["name"])
    return list1
```

```python
# Apply convert function to genres column and assign the result back
df["genres"] = df["genres"].apply(convert)
df["genres"]
```

```python
# Apply convert function to keywords column and assign the result back
df["keywords"] = df["keywords"].apply(convert)
df["keywords"]
```

```python
# df["cast"] = ast.literal_eval(df["cast"])
(df["cast"][0])
```

```python
def convert(obj):
    list1 = []
    counter = 0
    for i in ast.literal_eval(obj):
         if counter == 3:
              break
         else:
            list1.append(i["name"])
            counter += 1
    return list1
```

```python
df["cast"] = df["cast"].apply(convert)
df["cast"][0]
```

```python
def convert(obj):
    list1 = []
    for i in ast.literal_eval(obj):
         if i["job"] == "Director":
                list1.append(i["name"])
    return list1
```

```python
df["director"] = df["crew"].apply(convert)
df["director"]
```

```python
df["overview"] = df["overview"].apply(lambda x: str(x).split())

df["overview"][0]
```

```python
df["director"] = df["director"].apply(lambda x:[i.replace(" ","") for i in x])
df["cast"] = df["cast"].apply(lambda x:[i.replace(" ","") for i in x])
df["genres"] = df["genres"].apply(lambda x:[i.replace(" ","") for i in x])
df["keywords"] = df["keywords"].apply(lambda x:[i.replace(" ","") for i in x])
```

```python
df["tags"] = df["director"] + df["cast"] + df["genres"] + df["keywords"] 
```

```python
df 
```

```python
df = df[["id","tags","original_title"]]
df["original_title"] = df["original_title"].str.lower()
df["tags"] = df["tags"].apply(lambda x: [i.lower() for i in x])


df
```

```python
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
def stemm(text):
    y = []
    
    if isinstance(text, list):
        words = text
    else:
        words = text.split()
    
    for i in words:
        y.append(ps.stem(i))
        
    return " ".join(y)

df["tags"] = df["tags"].apply(stemm)
```

```python
df["tags"]
```

```python
from sklearn.feature_extraction.text import CountVectorizer

# ensure tags is a string, not list
# df["tags"] = df["tags"].apply(lambda x: " ".join(x) if isinstance(x, list) else "")

cv = CountVectorizer(
    max_features=5000,
    stop_words="english"
)

vectors = cv.fit_transform(df["tags"]).toarray()
vectors
```

```python
from sklearn.metrics.pairwise import cosine_similarity

similarity = cosine_similarity(vectors)
similarity
```

```python
def recommend(movie_name):
    movie_name = movie_name.lower()
    movie_index = df[df["original_title"] == movie_name.lower()].index()[0]
    distance = similarity[movie_index]
    recommended_movies = sorted(list(enumerate(distance[0])),reverse=True,key=lambda x:x[1])[1:6]

    for i in recommended_movies:
        print(df.iloc[i[0]].original_title)
```
